```
import re
import requests, bs4
import resource
from lxml import html
from urllib.request import Request, urlopen
from urllib.parse import unquote, quote_plus, urlencode
from xml.etree import ElementTree as ET
from datetime import datetime, timedelta, date

# Service URL
xmlUrl = 'https://open.neis.go.kr/hub/mealServiceDietInfo'
My_API_Key = unquote('524f225787cc47bda14b2df6ed4da0ef') ## 본인의 키를 공공데이터포털에서 받아서 입력(입력필요) -> 자신의 키로 바꿔도 됨 
#f=open("/content/gdrive/MyDrive/정보(2021)/2021-3 meal.csv",'w') ## 자신이 원하는 이름을 입력하기
f=open("/content/gdrive/MyDrive/정보(2021)/"+sfullname+"3월.csv",'w') ## 자신이 원하는 이름을 입력하기 
f.write("날짜"+","+"학교명"+","+"식사명"+","+"요리명"+","+"칼로리정보" +"\n") ##데이터 제목 파일에 입력하기 
no_num = re.compile('[^0-9]')


timestart=date(2020,3,2)    ##시작할 날짜 지정

while date(2021,3,31)>=timestart:
    dstart=timestart.strftime("%Y%m%d")
    queryParams = '?' + urlencode(
        {
            quote_plus('KEY') : My_API_Key,
            quote_plus('ATPT_OFCDC_SC_CODE') : "K10", #강원도 교육청 
            quote_plus('SD_SCHUL_CODE') : snumber, # 강원사대부고 번호
            quote_plus('MLSV_FROM_YMD') : dstart,
            quote_plus('MLSV_TO_YMD') : dstart
        }
    )
    response = requests.get(xmlUrl + queryParams).text
    #print(response) #응답 확인용 코드
    root = ET.fromstring(response)
    items = root.findall("row")
    for item in items:
      school = item.find('SCHUL_NM').text #학교이름
      mealn = item.find('MMEAL_SC_NM').text #식사명
      dishn = item.find('DDISH_NM').text #요리명 -> BR 태그 없애기
      dishn = dishn.replace('.','')
      dishn = "".join(no_num.findall(dishn))
      dishn = dishn.replace("<br/>",' ')
      #dishn = dishn.split("<br/>")
      cal = item.find('CAL_INFO').text
      cal = cal.replace('Kcal', '')
      dstart2=timestart
      dstart2=dstart2.strftime("%Y/%m/%d")
      f.write(dstart2+","+school+","+mealn+","+"".join(dishn)+","+cal+"\n")
      print(dstart2 + "기록 완료")
    timestart=timestart+timedelta(days=1)
f.close()
```

```
import matplotlib as mpl
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS #시각화를 위한 도구인 wordcloud(전체)에서 시각화를 담당하는 WordCloud도구와, 특정 단어를 제거할 수 있는 STOPWORDS 도구를 불러옴
import csv

f = open('/content/gdrive/MyDrive/정보(2021)/사대부고 급식 정보.txt','r') #* 파일 입력 클래스룸에서 다운받아 DPL에 집어놓은 moon.txt 파일 이름을 입력
f2 = open('/content/gdrive/MyDrive/정보(2021)/sdhs meal count.csv','w')
f2.write("메뉴"+","+"빈도"+"\n")

data = f.read()

sdhslist = WordCloud().process_text(data)
print(sdhslist)

for k,v in sdhslist.items():
  f2.write(k+","+str(v)+"\n")

#print(data)



spwords=set(STOPWORDS) # 기본적으로 많이 쓰는 단어 (제외할 단어) # ex)'should', "wouldn't", "how's", "i'm", 'then', etc ..
spwords.add('') #제외하고 싶은 단어 추가 * 시각화에서 큰 의미가 없는 단어들을 최소 3개 이상 채우기
spwords.add('') #제외하고 싶은 단어 추가
spwords.add('') #제외하고 싶은 단어 추가
spwords.add('') #제외하고 싶은 단어 추가


wordcloud = WordCloud(max_font_size=200, font_path='/content/gdrive/MyDrive/정보(2021)/강원교육모두+Light.ttf',    #폰트 크기, 폰트 경로, 제외할 단어, 배경색, 크기.generate(대상이 되는 글자)
                     stopwords=spwords,
                     colormap='Spectral',
                     background_color='#fffaf0',
                     width=1200,height=800).generate(data)

plt.figure(figsize=(20,10)) #크기 지정 
plt.imshow(wordcloud) #이미지 생성 
plt.tight_layout(pad=0) #여백 설정
plt.axis('off') #x,y축 값 삭제 
plt.savefig('/content/gdrive/MyDrive/정보(2021)/사대부고급식 2020-2021.png') #DPL에 moon.png로 파일 생성 
plt.show() #생성된 시각화 보여주기 
f.close() #파일 사용 마지막
f2.close()
```
