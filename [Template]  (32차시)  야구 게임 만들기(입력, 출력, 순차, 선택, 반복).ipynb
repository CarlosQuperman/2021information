### 중복 문제를 어떻게 해결할 수 있을까?

- 아무 숫자도 안 만든 상태에서는 숫자 하나 만들어서 그냥 저장
- 남은 2번의 숫자를 만들어야 함.
- 일단 숫자를 하나 만든다음에
- 그 숫자가 혹시 만들어진 숫자라면
- 중복되지 않는 숫자가 나올때까지 계속 생성 
- 생성이 끝나면 그 숫자를 리스트에 추가 시켜줌

```
import random #랜덤 숫자를 만들기 위해 필요한 도구 

base = [] #야구 게임에 사용될 숫자 3개를 저장할 리스트
ran = 0 #랜덤으로 만들 숫자를 임시로 저장할 변수

ran = random.randint(1,10) #1차적으로 숫자를 만들어 집어넣음
base.append(ran) #만들어진 숫자를 집어넣음


for i in range(2):  #1부터 10까지의 숫자를 랜덤으로 2번 생성함
  ran = random.randint(1,10) #일단 숫자를 만들어보고
  while (ran in base): #만든 숫자가 base에 있다면 다시 생성을 반복함
    ran = random.randint(1,10)
  base.append(ran) #중복없이 만들어진 숫자를 ran에 집어넣음

print(base) #생성된 숫자를 출력함
```

### 볼과 스트라이크 중에 어떤걸 먼저 처리해야 할까?

- 먼저 쉬운걸 처리하는게 훨씬 편하다.
- 볼과 스트라이크 중 어떤 걸 처리하는게 편할까?
- 볼이 편하다
- 볼을 먼저 처리해보자 (입력한 값이 만들어진 리스트에 있으면 볼임)
- in이라는 것은 어떤 원리로 동작할까?

```
import random #랜덤 숫자를 만들기 위해 필요한 도구 

base = [] #야구 게임에 사용될 숫자 3개를 저장할 리스트
ran = 0 #랜덤으로 만들 숫자를 임시로 저장할 변수
ball = 0 #볼의 개수를 저장할 변수 

ran = random.randint(1,10) #1차적으로 숫자를 만들어 집어넣음
base.append(ran) #만들어진 숫자를 집어넣음


for i in range(2):  #1부터 10까지의 숫자를 랜덤으로 2번 생성함
  ran = random.randint(1,10) #일단 숫자를 만들어보고
  while (ran in base): #만든 숫자가 base에 있다면 다시 생성을 반복함
    ran = random.randint(1,10)
  base.append(ran) #중복없이 만들어진 숫자를 ran에 집어넣음

print(base) #생성된 숫자를 출력함

a,b,c = map(int, input().split())

print(a,b,c)

if a  base: #a가 만들어진 숫자값인 base '안'에 있으면
  ball = ball + 1
if b  base:
  ball+=1
if c  base:
  ball+=1

print("볼은 ",ball,"개입니다.")
```
