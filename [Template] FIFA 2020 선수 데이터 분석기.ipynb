```
#선수들이 선호하는 발의 비율을 막대 그래프로(bar) 시각하

fifa2020['preferred_foot'].value_counts().plot(kind = '')
plt.legend() #범례 : 데이터가 무슨인지 설명해주는 조그만 표를 출력하게 함.
plt.savefig('') #경로를 설정하면 시각화 결과를 파일로 저장해줌 확장자는 .png
plt.show() #시각화 결과를 보여줌 

print(fifa2020['preferred_foot'].value_counts()) #선호하는 발 별로 정확한 개수를 보여줌
```

```
#선수들이 선호하는 발의 비율을 원 그래프(pie)로 시각화 (소수 둘째자리까지)

fifa2020['preferred_foot'].value_counts().plot(kind = '',autopct="%.f%%")
plt.legend()
plt.savefig('') #경로를 설정하면 시각화 결과를 파일로 저장해줌 확장자는 .png
plt.show()
```
