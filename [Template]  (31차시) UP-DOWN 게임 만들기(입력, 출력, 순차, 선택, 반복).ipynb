```
# (입력, 출력, 선택, 반복)
# 정답을 입력하는 기회를 입력받아 처리하도록  반복문의 조건을 채워보자
# 정답을 입력하는 기회는 n이라는 변수에 저장한다.

import random

a = int(input('아무 숫자나 입력하세요'))
ran = random.randint(1,a)
print(ran)

n = int(input('정답으로 인정할 회수를 입력하세요'))

for i in range(n):
  b = int(input('당신이 생각하는 정답을 입력하세요'))
  if b == ran:
   print('정답입니다')
  elif b <  ran:
   print('정답보다 더 작습니다')
  elif b > ran:
    print('정답보다 더 큽니다')
```

```
# (입력, 출력, 선택, 반복 종합)
# for문은 반복의 회수를 알고 있을 때 사용하고
# while문은 반복의 회수를 모를 때 사용한다. (정답을 몇번만에 맞출 수 있을지 알 수 없음 )
# while의 조건으로 b가 ran과 같지 않을때를 주고 
# 입력받을 때마다 count값을 1씩 증가시킨 후 
# count가 n이하이면 합격, 그렇지 않으면 불합격을 출력한다.
# print(ran) 앞에는 #을 써준다. 


count = 0
b = -1
a = int(input('아무 숫자나 입력하세요'))
ran = random.randint(1,a)
#print(ran)

n = int(input('정답으로 인정할 회수를 입력하세요'))

while(b!=ran):
  b = int(input('당신이 생각하는 정답을 입력하세요'))
  count = count + 1
  if b == ran:
   print('정답입니다')
  elif b < ran:
   print('정답보다 더 작습니다')
  elif b > ran:
    print('정답보다 더 큽니다')

print("정답 맞춘 회수",count)

if count <= n  :
  print("승리")
else:
  print("패배")

```

### 위 게임을 어떻게 수정할 수 있을까?

- 정답을 맞춘 회수에다가 추가로
- 정답과 내가 입력한 추측값의 거리를 점수로 추가
- 정답이 10이고 내가 입력한 값이 3이면 오차가 7 -> 10 - 7 = 3점부여
- 정답이 10이고 내가 입력한 값이 9이면 오차가 1 -> 10 - 1 = 9점부여
- 이외에도 자유자재로 수정할 수 있음
- 프로그래밍에서 랜덤값(임의의 값)을 생성하는 것은 매우 중요
- 주사위를 던지는 경우, 게임에서 확률을 반영하는 경우 등 많은 프로그래밍에 있어 랜덤값은 중요
- 그래서 위 프로그램과 학습한 내용을 바탕으로 아래에 자유자재로 프로그램을 작성해봅시다. (선택)
