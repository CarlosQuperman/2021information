* 바로 위 코드에서 여러 도형을 그리기 위해서 반복문을 직접 복붙했다.
* 그런데 일일히 도형을 그릴 때마다 복붙을 하기에는 너무 귀찮다.
* 각각의 반복문은 각각의 도형을 그리는 문제를 해결한다.
* 그 문제 해결방법에 이름을 붙여주고
* 그것을 사용하면 함수이다!
* 파이썬에서 함수는 def 함수이름 (입력으로)로 정의한다.

```python
def start(): #시작하는 함수
  import ColabTurtle.Turtle as t #도구 이름이 너무 기니 t로 줄여주기!
  t.initializeTurtle()
  t.bgcolor('skyblue') 
  t.color('blue') 
  t.speed(3)

def rect(): #(포트폴리오!!) ()을 그리는 함수 
  for i in range(4): 
    t.fd(120)
    t.rt(90)

def tri(): #(포트폴리오!!) ()을 그리는 함수 
  for i in range(3): 
    t.fd(120) 
    t.rt(120)

def star(): #(포트폴리오!!) ()을 그리는 함수 
  for i in range(1,6): 
    t.fd(100) 
    t.rt(144) 

```

```python
start()
rect()
tri()
star()

t.pu() #펜 들기 
t.goto(300,400) #x축 300, y축 400으로 거북이 이동
t.pd() #팬네리기 

star()
```

* 거북이의 위치를 지정하는 함수도 세줄이라 마음에 안든다!
* 함수로 만들어보자!
* 그런데 거북이의 위치를 지정하는 함수는 x와 y라는 두 개의 값을 지정해줘야 한다.
* 그렇다면 위치를 이동하는 데 2개의 입력이 필요하다
* 그러면 함수의 ()안에 입력의 이름과 개수를 넣어주면 된다

```python
 def move(x,y):
  t.pu()
  t.goto(x,y)
  t.pd() 
```

```python
start()
rect()
move(100,300)
tri()
move(10,10)
star()
```

* 함수를 사용하면 다시 **사용하기가(재사용) 편하다.** : 별 두 개 그리려면 star()를 2번 사용
* 함수를 사용하면 **내용을 수정하기가 편하다 :** 단순 복붙으로 3개의 별을 그렸는데 3개의 별의 각 변 길이를 바꿔야 한다면 3개를 다 바꿔야 한다.
* 그래서 함수 star()를 활용한다면 star()를 def 한 부분만 수정해주면 된다.
* 또 star에 변의 길이를 입력받아 사용하고 싶다면
* def star(size): 
  for i in range(1,6): 
    t.fd(size) 
    t.rt(144) 
* 괄호 안에 필요할 때마다 길이를 수정 해 별을 그릴 수 있다
* 함수를 사용하면 **여러 사람이 프로그램을 만들 수 있다**. : 하나의 프로그램을 하나의 프로그래머가 만들 수 있을까?
* 함수를 사용한다면 각각의 문제를 해결하는 프로그램을 각각의 프로그래머, 기획자, 팀이 만든 후 이를 합치면 된다.
* 함수에서 더 나아가면 클래스라는 개념이 있는데 이는 차후 기회가 있으면 ^^;

* 아래에 있는 내용은 코랩 터틀에서 사용할 수 있는 다양한 기능들이다.
* 그런데 함수의 공통점으로 ()가 있다는 것을 생각해보자.
* t.initializeTurtle(), t.shape("turtle")도 함수일까?
* 함수가 맞다.
* .이라는 것은 그 요소 안에 함수가 있다는 것을 의미하는데 
* 이를 **메서드**라고 한다!
* 함수는 문제를 해결하는 해결방법을 의미하고
* 메서드는 주체, 객체가 가지고 있는 해결방법을 의미한다!
* 걷기(), 달리기(), 밥먹기() <- 함수
* 장원재.걷기(), 장원재.달리기() <- 메서드

```python
# 코랩에서 안되는 것들은 앞에 # 붙여놓음

# 초기 세팅 
import ColabTurtle.Turtle as t# 라이브러리 불러오기
t.initializeTurtle() #거북이 만들기 
t.shape("turtle")      # 커서모양 arrow, circle, square, triangle, classic

# 설정함수 - 메서드 (배경색깔, 펜 색깔, 펜 모양)

t.bgcolor("black")  # red, purple, blue, green, yello, orange
t.bgcolor(r,g,b) #r,g,b 색깔을 입력하면 원하는 색깔 표시할 수 있음
t.pencolor("red")        # 선색상
t.pencolor(r,g,b) #r,g,b, 색깔을 입력하면 원하는 펜 색깔 표시할 수 있음
t.speed(0)               # 이동속도
t.width(3)               # 선굵기


#이동함수 - 메서드 
t.forward(100)  # 전진, 길이 = t.fd(100)
t.backward(100) # 후진, 길이 = t.bk(100)
t.right(90)     # 우회전, 각도 = t.rt(90)
t.left(90)      # 좌회전, 각도 = t.lt(90)
#t.circle(100)   # 원, 반지름 
t.penup()       # 펜 올리기, 그리지 않기 = t.up()
t.pendown()     # 펜 내리기, 그리기 = t.down()
t.goto(0,0)     # 좌표이동
t.heading()     # 현재의 방향을 반환
t.position()    # 현재의 위치 반환
t.dot()         # 현재의 위치에 점찍기

#화면에 글짜 출력

t.write("안녕하세요")

#색상 채우기  (코랩에서는 안됨 ㅠㅠ)

#t.fillcolor("yellow")  # 채우기 색상. red, blue, green
#t.begin_fill()       # 채우기 시작
#t.circle(100)
#t.end_fill()         # 채우기 종료

```

```python
#추가 예제 2
import ColabTurtle.Turtle as tree

tree.initializeTurtle()
tree.bgcolor(255,255,255) 
tree.speed(10)
MINIMUM_BRANCH_LENGTH = 5

def build_tree(t, branch_length, shorten_by, angle):
  if branch_length > MINIMUM_BRANCH_LENGTH:
    t.forward(branch_length)
    new_length = branch_length - shorten_by

    t.left(angle)
    build_tree(t, new_length, shorten_by, angle)

    t.right(angle * 2)
    build_tree(t, new_length, shorten_by, angle)

    t.left(angle)
    t.backward(branch_length)

tree.hideturtle()
tree.setheading(90)
tree.color('green')

build_tree(tree, 50, 5, 30)
```
