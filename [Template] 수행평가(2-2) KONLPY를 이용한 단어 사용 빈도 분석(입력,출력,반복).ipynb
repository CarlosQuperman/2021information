### 1. 내가 관심 있는 텍스트 품사 분석(KonlPy)

*   KonlPy를 사용해서 **저번 수행평가에 사용한 텍스트 또는 내가 관심 있는 또 다른 텍스트(한글 파일에 붙여넣기 했을 때 반 정도 이상 분량)를 정보-2021에 업로드**하고 
*  그 파일의 **경로를 복사하여 입력값을 설정해줍시다.**
* KonlPy는 대상이 되는 텍스트를 띄어쓰기 단위가 아닌(X) **단어 단위(O)**로 분석하여 단어와 품사의 짝이되게 출력합니다.
* **요구하는 모든 조건을 수행해야 정답으로 인정합니다.**


```
import csv  #분석 결과를 파일로 출력하기 위한 도구 
import konlpy #품사 분석을 위한 konlpy 불러옴 

f = open('','') ### 분석의 대상이 되는 텍스트의 경로를 첫번째 ' ' 사이에 넣고 목적을 두번째 ' ' 사이에 넣어줌

data = f.read() #f에 있는 모든 값들을 읽어와 data에 저장 
sentence = data #분석을 위해 sentence에 data값 저장 
words = konlpy.tag.Okt().pos() ###pos안에 분석의 대상이 되는 텍스트의 이름을 입력하고 분석값을 word에 집어넣음
print(words) #분석 결과값을 출력함

### word는 ('', '')의 형태로 저장된 리스트
### 단어는 몇번째? (       )
### 품사는 몇번째? (        )
### words안에 있는 값들을 단어와 품사로 나누어 저장한다면 분석 대상이 되는 텍스트와 그 품사를 표로 저장할 수 있다. 

f.close() #읽어온 파일 사용 마무리

```

# 2. 텍스트 품사 분석의 결과를 파일로 출력하기 

* 텍스트 분석의 결과를 좀 더 효과적으로 사용하려면 이를 나중에 활용하기 쉽게 파일로 저장할 필요가 있습니다.
* Konlpy의 분석결과인 words는 (단어, 품사)의 형태로 이루어져 있습니다.
* 그렇다면 **첫 줄에 단어, 품사라는 데이터의 제목을 넣고**
* 제목의 형태로 처음부터 끝까지 저장하여 파일로 만든다면
* 이를 쉽게 활용할 수 있습니다.
* **데이터를 읽어와 f에 저장하고**
* 그 데이터를 Konlpy로 분석한 **결과를 f2에 넣어주도록 경로와 목적을 설정해주고**
* **제일 첫행에 제목을 설정**해준 후 
* **의미 없는 값들의 조건을 설정하여 제외한 후 그 결과를 csv 파일로 만들어주는 파일 출력 프로그래밍**을 수행해봅시다.
